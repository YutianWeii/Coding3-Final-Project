### 1. Split the video, with each frame as an image.

```python
import cv2

def extract_frames(video_path, output_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)
    
    # Check if the video file is successfully opened
    if not video.isOpened():
        print("Unable to open the video file")
        return
    
    # Get the frame rate and total number of frames of the video
    fps = video.get(cv2.CAP_PROP_FPS)
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Video FPS: {fps}")
    print(f"Total frames: {total_frames}")
    
    # Loop through and save the frame images
    current_frame = 0
    while True:
        # Read a frame image
        ret, frame = video.read()
        
        # Check if the frame image is successfully read
        if not ret:
            break
        
        # Save the frame image
        frame_filename = f"{output_path}/frame_{current_frame}.jpg"
        cv2.imwrite(frame_filename, frame)
        
        # Display the current progress
        print(f"Processing frame {current_frame}/{total_frames}")
        
        # Increment the frame counter
        current_frame += 1
    
    # Release the video object
    video.release()
    print("Processing complete")

# Call the function to extract video frames
video_path = "/Users/apple/Desktop/UAL study/coding 3/Final/Ice Merchants.mp4"  # Path to the input video file
output_path = "/Users/apple/Desktop/UAL study/coding 3/Final/frames-IM"  # Path to the folder to save the frame images
extract_frames(video_path, output_path)
```

### 2. As the change between each frame and its adjacent frames is not very obvious, it has been adjusted to extract one image every 100 frames.

```python
import cv2

def extract_frames(video_path, output_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)
    
    # Check if the video file is successfully opened
    if not video.isOpened():
        print("Unable to open the video file")
        return
    
    # Get the frame rate and total number of frames of the video
    fps = video.get(cv2.CAP_PROP_FPS)
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Video FPS: {fps}")
    print(f"Total frames: {total_frames}")
    
    # Loop through and save the frame images
    current_frame = 0
    while True:
        # Read a frame image
        ret, frame = video.read()
        
        # Check if the frame image is successfully read
        if not ret:
            break
        
        # If current frame is a multiple of 100, save the frame image
        if current_frame % 100 == 0:
            frame_filename = f"{output_path}/frame_{current_frame}.jpg"
            cv2.imwrite(frame_filename, frame)
            
            # Display the current progress
            print(f"Processing frame {current_frame}/{total_frames}")
        
        # Increment the frame counter
        current_frame += 1
    
    # Release the video object
    video.release()
    print("Processing complete")

# Call the function to extract video frames
video_path = "/Users/apple/Desktop/UAL study/coding 3/Final/Ice Merchants.mp4"  # Path to the input video file
output_path = "/Users/apple/Desktop/UAL study/coding 3/Final/frames-IM2"  # Path to the folder to save the frame images
extract_frames(video_path, output_path)
```


### 3. To fit the pix2pix model setting, each frame is resized to 256 * 256 and then combined with its adjacent frame side by side to form a single image. The output image is 512 * 256 in size.


```python
import os
import cv2
import numpy as np

def combine_images(input_folder, output_folder):
    # Create the output folder
    os.makedirs(output_folder, exist_ok=True)
    
    # Get the list of files in the input folder
    files = os.listdir(input_folder)
    files.sort()  # Sort by file name
    
    # Loop through and combine adjacent images
    for i in range(len(files) - 1):
        # Read the adjacent two images
        img1 = cv2.imread(os.path.join(input_folder, files[i]))
        img2 = cv2.imread(os.path.join(input_folder, files[i+1]))
        
        # Resize the images to the target size
        img1 = cv2.resize(img1, (256, 256))
        img2 = cv2.resize(img2, (256, 256))
        
        # Create a new combined image
        combined_img = np.hstack((img1, img2))
        
        # Resize the combined image to the target size
        combined_img = cv2.resize(combined_img, (512, 256))
        
        # Generate the output file path
        output_filename = os.path.join(output_folder, f"combined_{i}.jpg")
        
        # Save the combined image
        cv2.imwrite(output_filename, combined_img)
        
        # Display the processing progress
        print(f"Processing {i+1}/{len(files)-1}")
    
    print("Processing complete")

# Call the function to combine images
input_folder = "/Users/apple/Desktop/UAL study/coding 3/Final/frames-IM2"  # Path to the input folder
output_folder = "/Users/apple/Desktop/UAL study/coding 3/Final/dataset-IM2"  # Path to the output folder to save the combined images
combine_images(input_folder, output_folder)
```


### 4. he images in the dataset are distributed into the training set, validation set, and test set in a ratio of 70%, 15%, and 15%, respectively.


```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Dataset path
dataset_dir = '/Users/apple/Desktop/UAL study/coding 3/Final/dataset-IM2'

# Outputpath for training, validation and test set
train_dir = '/Users/apple/Desktop/UAL study/coding 3/Final/train-IM2'
val_dir = '/Users/apple/Desktop/UAL study/coding 3/Final/val-IM2'
test_dir = '/Users/apple/Desktop/UAL study/coding 3/Final/test-IM2'

# Get all image filenames
image_filenames = os.listdir(dataset_dir)

# Use 70% of data as the training set, 30% as a temporary set (which will be split into validation and test set)
train_filenames, temp_filenames = train_test_split(image_filenames, test_size=0.3, random_state=42)

# Split the temporary set into validation set and test set, each 50% of temporary set
val_filenames, test_filenames = train_test_split(temp_filenames, test_size=0.5, random_state=42)

# Ensure the output directory exists
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Copy the images to the corresponding directories
for filename in train_filenames:
    shutil.copy(os.path.join(dataset_dir, filename), os.path.join(train_dir, filename))

for filename in val_filenames:
    shutil.copy(os.path.join(dataset_dir, filename), os.path.join(val_dir, filename))

for filename in test_filenames:
    shutil.copy(os.path.join(dataset_dir, filename), os.path.join(test_dir, filename))
```
